```python
#queue

import matplotlib.pyplot as plt

class Queue:
    def __init__(self, max_size=5):
        self.max_size = max_size
        self.front = -1
        self.rear = -1
        self.data = [None] * self.max_size

    def is_full(self):
        return (self.rear + 1) % self.max_size == self.front

    def is_empty(self):
        return self.front == -1

    def enqueue(self, value):
        if self.is_full():
            print("Maaf, queue penuh")
        else:
            if self.is_empty():
                self.front = self.rear = 0
            else:
                self.rear = (self.rear + 1) % self.max_size
            self.data[self.rear] = value
            print(f"Data {value} masuk ke queue")
            self.visualize()

    def dequeue(self):
        if self.is_empty():
            print("Data telah kosong!")
        else:
            print(f"Data yang terambil: {self.data[self.front]}")
            self.data[self.front] = None
            if self.front == self.rear:
                self.front = self.rear = -1
            else:
                self.front = (self.front + 1) % self.max_size
            self.visualize()

    def print_queue(self):
        if self.is_empty():
            print("Queue kosong")
        else:
            print("\nData yang terdapat dalam queue:")
            print("--------------------------------")
            i = self.front
            while True:
                print(self.data[i], end="    ")
                if i == self.rear:
                    break
                i = (i + 1) % self.max_size
            print()
            self.visualize()

    def clear(self):
        self.front = self.rear = -1
        self.data = [None] * self.max_size
        print("\nSekarang queue kosong")
        self.visualize()

    def visualize(self):
        fig, ax = plt.subplots()
        ax.set_xlim(-1, self.max_size)
        ax.set_ylim(-1, 2)
        for i in range(self.max_size):
            rect = plt.Rectangle((i, 0), 1, 1, edgecolor='black', facecolor='white')
            ax.add_patch(rect)
            if self.data[i] is not None:
                ax.text(i + 0.5, 0.5, str(self.data[i]), ha='center', va='center')
        ax.text(self.front, -0.5, 'Front', ha='center', va='center', color='red')
        ax.text(self.rear, 1.5, 'Rear', ha='center', va='center', color='blue')
        plt.axis('off')
        plt.show()

def main():
    queuebaru = Queue()
    while True:
        print("\tPROGRAM QUEUE")
        print("\t===============")
        print("Menu:")
        print("1. Dequeue")
        print("2. Enqueue")
        print("3. Cetak")
        print("4. Bersihkan queue")
        print("5. Exit")

        menu = input("Menu pilihan Anda: ")

        if menu == '1':
            queuebaru.dequeue()
        elif menu == '2':
           gatau = int(input('Masukan Jumlah Data Yang Akan Disimpan: '))
           if gatau > 5:
            print("Maaf, Jumlah Data Dalam Stack Tidak Boleh Lebih Dari 5")
            continue
           for i in range(gatau):
            dta = float(input(f"Data {i+1} yang akan disimpan di stack : "))
            queuebaru.enqueue(dta) #untuk masukkan data sesuai keinginan
        elif menu == '3':
            queuebaru.print_queue()
        elif menu == '4':
            queuebaru.clear()
        elif menu == '5':
            break
        else:
            print("Menu tidak valid")

        ulang = input("\n\nUlang? (y/t): ")
        if ulang.lower() != 'y':
            break

if __name__ == "__main__":
    main()
```

```python
class Stack:
    def __init__(self, max_size=5):
        self.max_size = max_size
        self.top = -1
        self.data = [0] * self.max_size

    def is_full(self):
        return self.top == self.max_size - 1

    def is_empty(self):
        return self.top == -1

    def push(self, value):
        if self.is_full():
            print("Maaf, stack penuh")
        else:
            self.top += 1
            self.data[self.top] = value

    def pop(self):
        if self.is_empty():
            print("Data telah kosong!")
            return None
        else:
            value = self.data[self.top]
            self.top -= 1
            return value

    def print_stack(self):
        print(f"Sisa stok: {self.top + 1}\n")

    def clear(self):
        self.top = -1
        print("\nSekarang stack kosong")


class Queue:
    def __init__(self, max_size=10):
        self.max_size = max_size
        self.front = 0
        self.rear = -1
        self.count = 0
        self.data = [None] * self.max_size

    def is_full(self):
        return self.count == self.max_size

    def is_empty(self):
        return self.count == 0

    def enqueue(self, value):
        if self.is_full():
            print("Antrian penuh!")
        else:
            self.rear = (self.rear + 1) % self.max_size
            self.data[self.rear] = value
            self.count += 1

    def dequeue(self):
        if self.is_empty():
            print("Antrian kosong!")
            return None
        else:
            value = self.data[self.front]
            self.front = (self.front + 1) % self.max_size
            self.count -= 1
            return value

    def print_queue(self):
        print("Antrian pengiriman:")
        if self.is_empty():
            print("Kosong")
        else:
            i = self.front
            for _ in range(self.count):
                print(self.data[i], end=" <- ")
                i = (i + 1) % self.max_size
        print()


def raja_cosplay_kemayoran():
    # Inisialisasi stok kostum dengan stack
    stok_kostum = {
        'R': Stack(5),
        'Y': Stack(5),
        'I': Stack(5),
        'S': Stack(5),
        'F': Stack(5)
    }

    # Mengisi stack dengan stok awal
    for key in stok_kostum.keys():
        for _ in range(5):
            stok_kostum[key].push(f"Stok {key}")

    # Inisialisasi antrian pengiriman
    queue_pengiriman = Queue()

    while True:
        print("\nRAJA COSPLAY KEMAYORAN")
        print("================================================================")
        print("1. Tambah Pelanggan")
        print("2. Proses Pengiriman")
        print("3. Tampilkan Antrian Pengiriman")
        print("4. Keluar")
        pilihan = input("Pilih opsi (1/2/3/4): ")

        if pilihan == '1':
            print("================================================================")
            print("INPUT DATA DIRI")
            nama = input("Masukkan nama: ")
            alamat = input("Masukkan alamat: ")

            print("================================================================")
            print("PILIHAN KOSTUM")
            jenis = [
                "R = Rimuru Tempest",
                "Y = Yukinoshita Yukino",
                "I = Ijichi Nijika",
                "S = Sunaookami Shiroko",
                "F = Firefly Star Rail"
            ]
            for i in jenis:
                print(i)

            kode = input("Masukkan kode kostum: ").upper()
            if kode not in stok_kostum:
                print("Kode tidak ditemukan")
                continue

            if stok_kostum[kode].is_empty():
                print(f"Stok kostum {kode} habis!")
                continue

            harga = 0
            print("S - L - M - XL")
            size = input("Pilih Size: ").upper()
            if kode == "R":
                harga = 50000 if size == "S" else 55000 if size == "M" else 60000 if size == "L" else 65000 if size == "XL" else 0
            elif kode == "Y":
                harga = 50000 if size == "S" else 60000 if size == "M" else 70000 if size == "L" else 80000 if size == "XL" else 0
            elif kode == "I":
                harga = 55000 if size == "S" else 65000 if size == "M" else 75000 if size == "L" else 85000 if size == "XL" else 0
            elif kode == "S":
                harga = 60000 if size == "S" else 65000 if size == "M" else 70000 if size == "L" else 75000 if size == "XL" else 0
            elif kode == "F":
                harga = 70000 if size == "S" else 75000 if size == "M" else 80000 if size == "L" else 85000 if size == "XL" else 0

            if harga == 0:
                print("Ukuran tidak valid")
                continue

            print(f"Anda memilih kostum {kode} ukuran {size}")
            stok_kostum[kode].pop()  # Mengurangi stok
            print("Total pembelian anda adalah= Rp.", harga)

            bayar = int(input("Masukkan pembayaran: Rp."))
            kembali = bayar - harga
            if bayar < harga:
                print("Uang anda tidak cukup")
                continue
            elif bayar >= harga:
                print("Kembalian anda Rp.", kembali)

            # Tambahkan pelanggan ke antrian pengiriman
            queue_pengiriman.enqueue((nama, alamat, kode, size))
            print(f"{nama} telah ditambahkan ke antrian pengiriman.")

        elif pilihan == '2':
            if queue_pengiriman.is_empty():
                print("Tidak ada pelanggan dalam antrian pengiriman!")
                continue

            pelanggan = queue_pengiriman.dequeue()
            print(f"\nMengirimkan kostum kepada {pelanggan[0]}")
            print(f"Alamat: {pelanggan[1]}")
            print(f"Kostum: {pelanggan[2]} ukuran {pelanggan[3]}")
            print("Pengiriman diproses...\n")

        elif pilihan == '3':
            queue_pengiriman.print_queue()

        elif pilihan == '4':
            print("\nTerima kasih sudah menggunakan layanan kami.")
            break

        else:
            print("Pilihan tidak valid, silakan coba lagi.")

        # Tampilkan sisa stok setiap selesai input
        print("\nSisa Stok Kostum:")
        for key in stok_kostum.keys():
            print(f"Kostum {key}:")
            stok_kostum[key].print_stack()

    print("\nAntrian Pengiriman:")
    queue_pengiriman.print_queue()


# Memanggil fungsi
if __name__ == "__main__":
    raja_cosplay_kemayoran()
```
